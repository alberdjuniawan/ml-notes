# Scikit-learn for Regression Models

```python
# import required library

import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
```

```python
# Load the diabetes dataset while directly separating and returning only X as the feature matrix and y as the target variable, 
# without returning the complete dataset object.

X, y = datasets.load_diabetes(return_X_y=True)
print(X.shape)
print(X[0])
```

```python
# Selecting one feature column by slice into [row, column], but [:, 2] returns a one-dimentional array. Therefore, we need to reshape it
# into two-dimentional array. This is necessary because scikit learn expects two-dimentional input (n_samples, n_features).

X = X[:, 2]
X = X.reshape((-1,1))
```

```python
# Check the first element of X after slicing and reshaping.

print(X[0])
```

```python
# Split the dataset into training and testing sets.
# test_size=0.33 means 33% of the data is used for testing

X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

```python
# Create linear regression model and train the model using the training data.

model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
# Make a prediction using the training data.

y_pred = model.predict(X_test)
```

```python
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel('Scalled BMIs')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes Progression Againts BMI')
plt.show()
```

### Case Study

Take a look at the Linnerud dataset in Scikit-learn. This dataset has multiple targets: 'It consists of three exercise (data) and three physiological (target) variables collected from twenty middle-aged men in a fitness club'.

In your own words, describe how to create a Regression model that would plot the relationship between the waistline and how many situps are accomplished. Do the same for the other datapoints in this dataset.

```python
from sklearn.datasets import load_linnerud

data = load_linnerud()

print(data.feature_names)
print(data.target_names)
```

```python
X = data.data
y = data.target

print(X.shape)
print(X[0])

print(y.shape)
print(y[0])
```

```python
X = X[:, 1:2]
y = y[:, 1:2]

print(X[0])
print(y[0])
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel('Situps')
plt.ylabel('Waist')
plt.title('Relationship between the waistline and how many situps are accomplished')
plt.show()
```

Explanation:

To create a regression model that plots the relationship between waist circumference and the number of sit-ups performed, the dataset must first be sliced to select one feature (sit-ups) and one target variable (waist). The dataset is then divided into training and testing sets. A linear regression model is built and trained using the training data. After the model has been fitted, predictions are generated using the testing set. Finally, the results are visualized by plotting the regression line together with the observed data points.
